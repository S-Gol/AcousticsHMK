#### Problem 2.4


A loudspeaker has a free cone resonance frequency of 55.5 Hz. If a
US nickel (mass = 5.00 g) is placed on the speaker’s dust cap, the resonance frequency is reduced to
46.3 Hz. After the nickel is removed, the speaker is installed in a sealed enclosure, and the
frequency increases to 80.0 Hz. What is the equivalent stiffness (in Newtons/meter) provided by
the elasticity of the air contained within the loudspeaker enclosure?



Assume loudspeaker can be treated as a mass on a spring. 

Assume air equivalent stiffness is in "parallel" to the loudspeaker stiffness. 

| Scenario | Mass, g | Freq, hz | Addtl Stiffness | 
|-----|-----|-----|-----|
| Base, 0 | $ M_0 $ | 55.5 | 0 |
| Nickel, 1 | $ M_0 $ + $ M_n $ | 46.3 | 0 |
| Air, 2 | $M_0 $ | 80 | $ k_a $ |



$f=\frac{1}{2\pi}\sqrt{\frac{k}{m}}$

{0} $55.5hz=f_0=\frac{1}{2\pi}\sqrt{\frac{k_0}{m_0}}$

{1} $46.3hz=f_1=\frac{1}{2\pi}\sqrt{\frac{k_0}{m_0+m_n}}$

{2} $80.0hz=f_2=\frac{1}{2\pi}\sqrt{\frac{k_0+k_a}{m_0}}$

Solving {0}, {1}, and {2} simultaneously 

$k_0 = \frac{4\pi^2m_n f_0^2 f_1^2}{f_0^2-f_1^2}$

$m_0=\frac{m_n f_1^2}{f_0^2-f_1^2}$

$k_a=\frac{m_nf_1^2 4 \pi^2 (f_2^2-f_0^2)}{f_0^2-f_1^2}$



$k_a=1500N/m$

$k_0 = 1391N/m$

$m_0 =0.11kg$

#### Problem 2.5

Spar buoy oscillations. An oceanographic instrument package is placed at the bottom of a spar
buoy shown in Fig. 2.35. The mass (in air) of the entire buoy (instruments, batteries, tube, and
antenna) is 150 kg. The tubular portion (5 m long and 25 cm diameter) is partially submerged. You
may assume the density of the water, ρ = 1000 kg/m3, and the acceleration due to gravity, g = 9.8 m/s2.

(a) Equilibrium depth. How far from the surface of the water is the top of the 5.0 m long tubular
section when it is in equilibrium (at rest)?

(b) Frequency. What is the natural frequency of vibration of the buoy if it is displaced (vertically)
from its equilibrium position and released?



##### A

$f_b =V\rho= Ah\rho = 150kg$

$A=0.049m^2$

$h=\frac{150kg}{0.049m^2*1000\frac{kg}{m^3}}=3.1m$

##### B

$F(h)=-Ah\rho*g + 150kg*g$

$h''=\frac{F(h)}{m}$

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp


A=0.049
rho=1000
g=9.8
m=150
dt=0.1
tMax=10

#Numeric setup
def hpp(h):
    return (m*g-A*h*rho*g)/m


tRange=np.linspace(0,tMax,int(tMax/dt))
v=0
h=0
hRange=[]

#True result
t,c,k=sp.symbols("t c k")
x=sp.symbols("x",cls=sp.Function)

expr=sp.Eq(x(t).diff(t,t),m*g-A*x(t)*rho*g)

trueEqn=sp.dsolve(expr)
print(trueEqn)

for t in tRange:
    v+=hpp(h)*dt
    h+=v*dt
    hRange.append(h)
    
```

```python
plt.plot(tRange,hRange)
```

Since $\omega=21.92$, $f=3.48hz$, which matches the chart
