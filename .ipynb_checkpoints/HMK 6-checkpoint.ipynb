### 5.3

L0 = 5cm

f0 = 784

Sweep range of 392-784, find lengths

$f_1=\frac{\pi c_b \kappa}{8L^2}*1.1937^2$

$\frac{\pi c_b \kappa}{8} *1.1937^2= b$ is constant due to material and geometric properties

$f=\frac{b}{L^2}\implies L=\sqrt{\frac{b}{f}}$

```python
import numpy as np
import matplotlib.pyplot as plt

f0=784
L0=5/100 #cm to m

b=f0*L0**2

fRange=np.linspace(392,784)
lRange=np.sqrt(b/fRange)*100

fig,ax=plt.subplots()

ax.set_xlabel("Freq, hz")
ax.set_ylabel("Length, cm")

ax.plot(fRange, lRange)
```

### 15.2

A 19.2m circular waveguide with 2 compression drivers, terminated by anechoic cone.  

3 flush microphones at x=0, 3.17, 17.9m. 

Dry air at 100kPa, c0=345m/s, k=0.02587w/mk

f=880hz

Find:

Attenuation length (13.78)

Shock inception distance Gol'dberg Number for p=100, 300, 1000 Pa.

$\alpha_{tv}=\frac{1}{c\tau_{tv}}=-\frac{\delta_v+\frac{\gamma-1}{\gamma}\delta_k}{\alpha}\frac{\omega}{c}$

$\alpha = \frac{K}{\rho c_p}$

$\tau_{tv}=-\frac{E}{\dot {E}}$

$\delta_k=\sqrt{\frac{2k}{2 \pi f \rho c_p }}$

$\delta_v=\sqrt{\frac{\nu}{\pi f}}$



```python
xs=[0,3.17, 17.9]
p=100e3
c0=345
k=0.02587
rho = 1.225
nu=14.88e-6
cp=1.006e3
cv=0.7171e3
gamma=cp/cv
c=345
f=880
omega = 2*np.pi*f

alpha=k/(rho*cp)

dk = np.sqrt(2*k/(2*np.pi*f*rho*cp))
dv = np.sqrt(nu/(np.pi*f))

alpha_tv=-((dv+dk*(gamma-1)/gamma)/alpha)*omega/c
print(alpha_tv)
```

```python
ps=[100,300,1000]

for p in ps:
    print("p = "+str(p))
    M=p/(rho*c**2)
    wl = c/f

    Ds=5*wl/(12*np.pi*M)
    print("Ds: " + str(Ds))

    G=1/(alpha*Ds)
    print("G: " + str(G))
```

$C_n=\frac{2J_n(n\sigma)}{n\sigma}$, $\sigma = xDs$

```python
from scipy.special import jv

fs=[880, 1.76e3, 2.64e3,3.52e3]

for n in range(len(fs)):
    for p in ps:
        for x in xs[1:3]:
            mode=n+1
            M=p/(rho*c**2)
            wl = c/fs[n]
            Ds=5*wl/(12*np.pi*M)
            sigma = x/Ds
            cn=2*jv(mode, mode*sigma)/(mode*sigma)
            print("Mode {0}, Pressure {1}, distance {2}, Cn = {3}".format(str(mode), str(p), str(x), str(cn)))
```
